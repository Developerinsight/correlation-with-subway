```python
rain = {}
avg_temp = {}
daily_temp_diff = {}

for val in list(subset['역명'].unique()):
  station = subset.loc[subset['역명']==f'{val}', :]
  corr = station.corr()
  rain[val] = corr.loc[['강수량(mm)'],  ['승차총승객수', '하차총승객수']]
  avg_temp[val] = corr.loc[['평균기온(℃)'],  ['승차총승객수', '하차총승객수']]
  daily_temp_diff[val] = corr.loc[['일교차'],  ['승차총승객수', '하차총승객수']]

```

```python
# 데이터를 그룹화합니다.
import seaborn as sns

grouped_data = subset.groupby('역명')

# 그룹별 상관관계 히트맵을 그리는 함수
def draw_correlation_heatmap(group):
    # 그룹 내 데이터에서 필요한 열을 선택합니다.
    #selected_columns = ['열1', '열2', …]  # 그룹 내에서 상관관계를 계산할 열 선택
    group_data = group#[selected_columns]
    
    # 상관관계 행렬을 계산합니다.
    corr_matrix = group_data.corr()
    
    # 상관관계 히트맵을 그립니다.
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap - Group: {}'.format(group.name))
    plt.show()

# 각 그룹에 대해 상관관계 히트맵을 그립니다.
grouped_data.apply(draw_correlation_heatmap)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

distance = mg3['거리']

# 거리 범주 나누기
distance_min = distance.min()
distance_max = distance.max()
distance_range = distance_max - distance_min
bin_size = distance_range / 3
categories = np.digitize(distance, bins=[distance_min + bin_size, distance_min + 2 * bin_size])

# 거리 카테고리별로 그래프 그리기
for category in range(0, 4):
    category_data = mg3[categories == category]
    plt.scatter(category_data['사용일자'], category_data['승차총승객수'], s=1)
    plt.xlabel('사용일자')
    plt.ylabel('승차총승객수')
    plt.title(f'거리 카테고리 {category}')
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gcf().autofmt_xdate()
    plt.show()
```
