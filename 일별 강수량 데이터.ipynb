```python
import pandas as pd
df4 = pd.read_csv('월별 다양한 기후자료.csv',encoding='cp949') #데이터를 불러온다
df4.rename(columns={'1월':0, '2월':1 ,'3월':2,'4월':3,'5월':4,'6월':5,'7월':6,'8월':7,'9월':8,'10월':9,'11월':10,'12월':11,'연합계':12},inplace=True)
df4 = df4.transpose()	#행 열 전환
df4.rename(columns=df4.iloc[0], inplace=True)	# 행열이 전환된 데이터프레임의 열 이름 제대로 수정
df4 = df4.drop(df4.index[0])
df4.head(13)
```

```python
df4= df4.astype(float)
df4.corr()
```

```python
import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
plt.style.use('default') # 스타일 서식 지정
plt.rc('font', family='Malgun Gothic')# 한글 폰트 지정
plt.rcParams['figure.figsize'] = [10, 8]
plt.pcolor(df4)

plt.xticks(np.arange(0.5, len(df4.columns), 1), df4.columns)

plt.yticks(np.arange(0.5, len(df4.index), 1), df4.index)

plt.title('Heatmap by plt.pcolor()', fontsize=20)

plt.xlabel('Year', fontsize=14)

plt.ylabel('Month', fontsize=14)

plt.colorbar()



plt.show()
```

```python
sns.heatmap(df4.corr(), annot=True, fmt='f')

plt.title('월별 데이터', fontsize=20)

plt.show()
```
