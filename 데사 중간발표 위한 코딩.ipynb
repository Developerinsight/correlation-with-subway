```python
import pandas as pd
df = pd.read_csv('/content/CARD_SUBWAY_MONTH_2022.csv', encoding='cp949')
df
```

```python
excluded_stations = ['신정네거리', '양천구청', '도림천', '신설동', '용두(동대문구청)', '신답', '용답']
train_data = train_data[~train_data['역명'].isin(excluded_stations)]
```

```python
import matplotlib.pyplot as plt # 맷플롯립 import 하기
plt.rcParams['font.family'] = 'NanumBarunGothic' # 나눔바른고딕 적용하

data = ['구로디지털단지', '신대방', '신림', '봉천', '서울대입구(관악구청)', '낙성대(강감찬)', '사당',
       '방배', '서초', '교대(법원.검찰청)', '강남', '역삼', '선릉', '삼성(무역센터)', '종합운동장',
       '잠실새내', '잠실(송파구청)', '잠실나루', '당산', '영등포구청',
       '문래', '신도림', '대림(구로구청)']
rain = {}
gyulbing = {}
lightning = {}
suri = {}
fog = {}
coldwave = {}
avgtemp = {}
holiday = {}

for val in list(train_data['역명']):
  station = train_data.loc[train_data['역명']==f'{val}', :]
  corr = station.corr()
  rain[val] = corr.loc[['강수일수'],  ['승차총승객수', '하차총승객수']]
  gyulbing[val] = corr.loc[['결빙일수'],  ['승차총승객수', '하차총승객수']]
  lightning[val] = corr.loc[['뇌전일수'],  ['승차총승객수', '하차총승객수']]
  suri[val] = corr.loc[['서리일수'],  ['승차총승객수', '하차총승객수']]
  fog[val] = corr.loc[['안개일수'],  ['승차총승객수', '하차총승객수']]
  coldwave[val] = corr.loc[['한파일수'],  ['승차총승객수', '하차총승객수']]
  avgtemp[val] = corr.loc[['평균기온'],  ['승차총승객수', '하차총승객수']]
  holiday[val] = corr.loc[['공휴일 수 '],  ['승차총승객수', '하차총승객수']]
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumBarunGothic'
plt.rc("axes", unicode_minus=False)

stations = list(gyulbing.keys())
total_get_on_values = [gyulbing[station]['승차총승객수']['결빙일수'] for station in stations]
total_get_off_values = [gyulbing[station]['하차총승객수']['결빙일수'] for station in stations]

plt.figure(figsize=(17, 7))

# Calculate the width of each bar
bar_width = 0.35

# Create a list of indices for the x-axis positions
indices = range(len(stations))

# Plot the "승차총승객수" as the first set of bars
plt.bar(indices, total_get_on_values, label='승차총승객수', width=bar_width)

# Shift the x-axis positions for the second set of bars
indices_shifted = [index + bar_width for index in indices]

# Plot the "하차총승객수" as the second set of bars
plt.bar(indices_shifted, total_get_off_values, label='하차총승객수', width=bar_width)

plt.xlabel('역명')
plt.ylabel('상관관계')
plt.xticks(indices_shifted, stations, rotation=90)
plt.legend(loc='best')
plt.grid(True)
plt.title('결빙일수')

plt.legend()
plt.show()
```
