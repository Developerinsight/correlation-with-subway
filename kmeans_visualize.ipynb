```python
import folium

def display_points_on_map(points):
    # 특정 위도, 경도 중심으로 하는 OpenStreetMap을 출력
    map_osm = folium.Map(location=[37.529622, 126.984307], zoom_start=12)

    # 마커를 그룹화할 FeatureGroup 생성
    marker_group = folium.FeatureGroup(name='Points')
    colors = ['blue', 'green', 'red', 'purple']

    # 각 점에 대한 마커 생성 및 FeatureGroup에 추가
    for i, point in points.iterrows():
        lat, lon = point['위도'], point['경도']
        color = colors[i % len(colors)]  # 마커 색상 선택
        marker = folium.Marker(location=[lat, lon], icon=folium.Icon(color=color))
        marker_group.add_child(marker)

    # FeatureGroup을 지도에 추가
    map_osm.add_child(marker_group)

    # 지도 출력
    return map_osm

# 점들의 위도, 경도 DataFrame
points = df_new3

# 점들을 지도에 표시
map_with_points = display_points_on_map(points)

# 지도 출력
map_with_points
```
